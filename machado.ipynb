# Palavras de Machado de Assis

Este *iPython Notebook* é um experimento para extrair as palavras do único texto de Machado de Assis disponível no Project Gutemberg: [Histórias sem data](https://www.gutenberg.org/ebooks/33056).

```python
original = open('txt/historias-sem-data.txt', encoding='utf-8').read()
len(original)
```

```python
partes = [palavra.lower() for palavra in original.split()]
len(set(partes))
```

```python
import collections
ct = collections.Counter(partes)
ct.most_common()
```
